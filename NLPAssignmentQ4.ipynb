Q4) Write a program to find all words that occur at least three times in the  BrownCorpus.(Hint : Use NLTK  corpus )

```python
import nltk
from nltk.corpus import brown
nltk.download('brown')
```

```python
all_unique_brown=set(brown.words())
brown_fd=nltk.FreqDist(brown.words())
atleast_3_times=[word for word in all_unique_brown if brown_fd[word]>2]
print(atleast_3_times)
```
